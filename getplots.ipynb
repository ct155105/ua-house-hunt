```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
full_df = pd.read_csv('Parcel_Boundaries.csv')
ua_df = full_df[full_df['CVTTXCD'] == 70]
```

```python
ua_df = full_df[full_df['CVTTXCD'] == 70]
```

```python
ua_df_resi = ua_df[ua_df['BLDTYP'].str.contains('Dwelling',na=False)]
```

```python
ua_df_resi_filtered = ua_df_resi[ua_df_resi['ShapeSTArea'] < 25000]
```

```python
np_array_x = ua_df_resi_filtered['ShapeSTArea'].to_numpy()
np_array_y = ua_df_resi_filtered['TOTVALUEBASE'].to_numpy()
```

```python
plt.scatter(np_array_x, np_array_y, 1)
```

```python
x = np_array_x.reshape((-1,1))
y = np_array_y
```

```python
model = LinearRegression().fit(x, y)
```

```python
model.coef_
```

```python
print(x)
```

```python
test_y = model.predict(x)
```

```python
plt.scatter(np_array_x, np_array_y, 1)
plt.plot(x,test_y, color='black', linewidth=3)
plt.show()
```
